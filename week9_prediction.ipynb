```python
# Import pandas for handling data in DataFrame format
import pandas as pd

# Import LinearRegression model from sklearn
# This model is used to predict continuous numeric values
from sklearn.linear_model import LinearRegression

# Create input data (features)
# Here, we are using age as the independent variable (X)
X = pd.DataFrame({'age': [20, 30, 40, 50, 60]})

# Create target values (what we want to predict)
# These could represent something like blood pressure or income
y = [110, 115, 120, 125, 130]

# Create the Linear Regression model
lr = LinearRegression()

# Train the model using the data
# The model learns the relationship between age (X) and y
lr.fit(X, y)

# New data we want predictions for
ages_to_predict = pd.DataFrame({'age': [10, 45, 80]})

# Use the trained model to make predictions
predictions = lr.predict(ages_to_predict)

# Print predicted values
print(predictions)
```

```python
# Loop through predictions and format them to 2 decimal places
# This makes the output easier to read
for p in predictions:
    print(f"{p:.2f}")
```

```python
# Using linear regression for binary outcomes (0 or 1)
# This is NOT ideal because predictions can go below 0 or above 1

X = pd.DataFrame({'age': [20, 30, 40, 50, 60, 70]})
y = [0, 0, 0, 1, 1, 1]

lr = LinearRegression()
lr.fit(X, y)

ages_to_predict = pd.DataFrame({'age': [5, 45, 100]})
predictions = lr.predict(ages_to_predict)

# These predictions can be negative or greater than 1 → not realistic
for p in predictions:
    print(f"{p:.2f}")
```

```python
# Import Logistic Regression
# This model is used for binary classification (0 or 1 outcomes)
from sklearn.linear_model import LogisticRegression
import numpy as np

# Input data
X = pd.DataFrame({'age': [20, 30, 40, 50, 60, 70]})
y = [0, 0, 0, 1, 1, 1]

# Create model
lr = LogisticRegression()

# Train model
lr.fit(X, y)

# New data
ages_to_predict = pd.DataFrame({'age': [5, 45, 100]})

# Get probabilities instead of raw predictions
# predict_proba returns probability of each class (0 and 1)
probabilities = lr.predict_proba(ages_to_predict)

# Print probability of class 1 (e.g., having disease)
for p in probabilities[:, 1]:
    print(f"{p:.2f}")
```

```python
# Import Decision Tree Classifier
# This model makes decisions based on splitting data into branches
from sklearn.tree import DecisionTreeClassifier

# Input data
X = pd.DataFrame({'age': [20, 30, 40, 50, 60, 70, 80]})
y = [0, 0, 1, 1, 1, 0, 0]

# Create model
dt = DecisionTreeClassifier()

# Train model
dt.fit(X, y)

# Predict new values
ages_to_predict = pd.DataFrame({'age': [25, 50, 75]})
probabilities = dt.predict(ages_to_predict)

# Print predictions
for p in probabilities:
    print(f"{p:.2f}")
```

```python
# This code visualizes how the decision tree makes decisions
# It shows how the model splits data based on age

from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))

plot_tree(dt,
          feature_names=['age'],
          class_names=['No Diabetes', 'Diabetes'],
          filled=True,
          rounded=True,
          fontsize=12,
          impurity=False,
          precision=0)

plt.title('Decision Tree — Age vs Diabetes')
plt.tight_layout()
plt.show()
```
